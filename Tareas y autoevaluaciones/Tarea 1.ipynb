```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

kepler = pd.read_csv('oec.csv')
pd.set_option('display.max_columns', None)
#pd.reset_option('max_columns')
kepler.head()
```

Acá vendrán los planetas asignados en diferentes grupos, donde cada grupo corresponde a un único método de descubrimiento.
Se mostrarán tres planetas por grupo para más comodidad pero se puede manipular la cantidad que se desea ver agregando el valor que guste dentro de head( )

Los grupos designados con las variables 'a', 'b', 'c', 'd' y 'e' corresponden a planetas descubiertos por el método de transit, RV, imaging, microlensing y timing respectivamente.





```python
a = kepler.groupby(['DiscoveryMethod']).get_group('transit')

a.head(3)
```

```python
b = kepler.groupby(['DiscoveryMethod']).get_group('RV')

b.head(3)
```

```python
c = kepler.groupby(['DiscoveryMethod']).get_group('imaging')

c.head(3)
```

```python
d = kepler.groupby(['DiscoveryMethod']).get_group('microlensing')

d.head(3)
```

```python
e=kepler.groupby('DiscoveryMethod').get_group('timing')

e.head(3)
```

Ahora usando la función agg() tomaremos las masas de los planetas y las sumaremos, esto por cada uno de los grupos:

Transit:

```python
a1 = a.agg('PlanetaryMassJpt').sum()
a1
```

RV:

```python
b1 = b.agg('PlanetaryMassJpt').sum()
b1
```

Imaging:

```python
c1 = c.agg('PlanetaryMassJpt').sum()
c1
```

Microlensing:

```python
d1 = d.agg('PlanetaryMassJpt').sum()
d1
```

Timing:

```python
e1 = e.agg('PlanetaryMassJpt').sum()
e1
```

Ahora haremos gráfico de barras ploteando en el eje x los Métodos de descubrimiento y en el eje y el valor de las masas sumadas por cada método

```python
x = np.array(['transit', 'RV', 'imaging', 'microlensing', 'timing'])
y = np.array([a1, b1, c1, d1, e1])

plt.bar(x, y)
```

Como extra me dio curiosidad saber cuantos valores tiene cada uno de los grupos, y compararlo con las barras del gráfico:

```python
kepler['DiscoveryMethod'].value_counts()
```

Como veo que hay más valores en el grupo 'Transit' que en el de 'RV', siendo este el que tiene una suma mayor, me dio curiosidad por saber como se distribuyen los valores de masa por cada grupo: 

```python
kepler.groupby('DiscoveryMethod')['PlanetaryMassJpt'].value_counts()
```

Al revisar de forma rápida se nota que hay aprox. igual o más valores númericos para RV que para transit, a pesar de que Transit es quién se registra más veces en la columna de Métodos de descubrimiento. En conclusión esto muestra que hay una gran cantidad de datos faltantes de masa con el método de Transit
