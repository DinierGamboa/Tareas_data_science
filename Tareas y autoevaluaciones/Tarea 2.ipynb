### Empezamos por importar los paquetes necesarios:

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Llamamos al archvo csv y lo visualizamos:

```python
met = pd.read_csv("../Sesión 3/meteorite-landings.csv")
#pd.set_option("display.max_row", None)
met.head(10)
```

### Ahora defininos un nuevo DataFrame llamado meteoritos_h4 el cuál solo tendrá meteoritos de la clase H4:

```python
meteoritos_h4 = met[(met.recclass.str.startswith("H4")) & (met.recclass.str.endswith("H4"))]
meteoritos_h4.head(15)
```

### También lo podemos visualizar usando rexp:

```python
meteoritos_h41 = met.recclass.str.findall('^H.*4$')
meteoritos_h41.head(15)
```

### Procedemos a hacer otro DataFrame llamado meteoritos_h5 para la clase H5:

```python
meteoritos_h5 = met[(met.recclass.str.startswith("H5")) & (met.recclass.str.endswith("H5"))]
meteoritos_h5.head(15)
```

### Una vez tenemos ambos DataFrames, queremos saber si poseen valores nulos en sus columnas de masa, por los que hacemos un conteo de valores: True para vacíos, y False para valores númericos:

```python
meteoritos_h4.mass.isnull().value_counts()
```

```python
meteoritos_h5.mass.isnull().value_counts()
```

### Podemos ver que el DataFrame de la clase H5 tiene 4 valores nulos, por lo que los eliminamos:

```python
meteoritos_h5 = meteoritos_h5.dropna(subset = ['mass'])
meteoritos_h5.head()
```

### Volvemos a contarlos y vemos que ya se ha limpiado la lista:

```python
meteoritos_h5.mass.isnull().value_counts()
```

### Ahora vamos a definir el DataFrame yrange4 que incluye a los meteoritos desde 1960 a 1990

```python
year_rangeH4 = meteoritos_h4[(meteoritos_h4.year > 1960) & (meteoritos_h4.year < 1990)]
yrang4 = year_rangeH4.reset_index()
yrang4.head()
```

### Revisamos y creamos un DataFrame con la cantidad de entradas para meteoritos por año en H4, en orden cronológico:

```python
metperyear4 =pd.DataFrame(year_rangeH4.year.value_counts())
mpy1 = metperyear4.reset_index() #Pasa la columna de año que estaba como columna índice a columna normal, y agrega una nueva como index
mpy1 = mpy1.rename(columns = {'index' : 'year', 'year': 'Nmeteor'}) # cambio los nombres de las columnas
mpy1.year = mpy1.year.astype(int) #cambio el valor float de los años a int
mpy1.sort_values(by = 'year', inplace= True) # se ordena por orden cronológico
mpy4 = mpy1.reset_index(drop = True) # Vuelvo a asignar indices en orden, y elimino los indices anteriores
mpy4.head()
```

### También vamos a necesitar saber la cantidad total de meteoritos para el DataFrame mpy4:

```python
mpy4.Nmeteor.sum()
```

### Ahora definimos el DataFrame yrang5 que incluye a los meteoritos desde 1960 a 1990:

```python
year_rangeH5 = meteoritos_h5[(meteoritos_h5.year > 1960) & (meteoritos_h5.year < 1990)]
yrang5 = year_rangeH5.reset_index()
yrang5.head()
```

### Revisamos y creamos nuevamente un DataFrame para la cantidad de meteoritos por año en H5, ordenada cronológicamente:

```python
metperyear =pd.DataFrame(year_rangeH5.year.value_counts())
mpy2 = metperyear.reset_index()
mpy2 = mpy2.rename(columns = {'index' : 'year', 'year': 'Nmeteor'})
mpy2.year = mpy2.year.astype(int) 
mpy2.sort_values(by = 'year', inplace= True)
mpy5 = mpy2.reset_index(drop = True) 
mpy5.head()
```

### Y observamos la cantidad total de meteoritos entre 1960-1990 para la clase H5:

```python
mpy5.Nmeteor.sum()
```

### Una vez tenemos los DataFrame yH4 y yH5 con los mismos indices los podemos unir en un único DataFrame (después del índice 2191 los datos relacionado a H4 serán nulos como es de esperar)

```python
full = pd.concat([yrang4, yrang5], axis=1, join="outer").head()
full.head()
```

### finalmente ploteamos la cantidad de meteoritos por año:

```python
plt.style.use('seaborn-whitegrid')

x1= mpy4.year
y1= mpy4.Nmeteor

col1 = []

for val in y1:
    if val >= 500:
        col1.append('red')
    elif val < 100:
        col1.append('green')
    else:
        col1.append('orange')

plt.bar(x1, y1, color = col1);
plt.suptitle('H4 class meteorites per year')
plt.ylabel('number of meteorites')
plt.show()
```

```python
plt.style.use('seaborn-whitegrid')

x2 = mpy5.year
y2 = mpy5.Nmeteor

col2 =[]

for val in y2:
    if val >= 300:
        col2.append('red')
    elif val < 100:
        col2.append('green')
    else:
        col2.append('orange')
        

plt.bar(x2, y2, color = col2)
plt.suptitle('H5 class meteorites per year')
plt.ylabel('number of meteorites')

plt.show()
```

### Y ploteamos la cantidad de meteoritos por clase:

```python
plt.style.use('seaborn-whitegrid')

x3 = np.array(['H4', 'H5'])
y3 = np.array([2192, 2633])

plt.bar(x3, y3, color = ['turquoise', 'mediumspringgreen'], edgecolor = 'darkslategray', width = 0.5)
plt.suptitle('meteorites per class')
plt.xlabel('class')
plt.ylabel('number of meteorites')
plt.show()
```
